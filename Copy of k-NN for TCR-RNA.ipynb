```python
z_mean = np.load("TCR-RNA Data/z_mean.npy")
specificity = np.load("TCR-RNA Data/specificity_resampled.npy")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(z_mean, specificity, test_size=0.2, random_state=42)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, KFold

# Assume we have training data X_train and corresponding labels y_train

# Define the list of k values to test
k_list = [i for i in range(1, 11)]

# Initialize KFold cross-validator
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize KNN classifier
knn = KNeighborsClassifier()

# Store the cross-validation scores for each k value
cv_scores = []

# Iterate through each k value
for k in k_list:
    # Set the k value of the KNN classifier
    knn.set_params(n_neighbors=k)
    
    # Perform cross-validation under the current k value and calculate the average score
    scores = cross_val_score(knn, X_train, y_train, cv=kf)
    cv_scores.append(scores.mean())
    
# Select the best k value
best_k = k_list[np.argmax(cv_scores)]
print("Best k value:", best_k)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Assume we have training data X_train and corresponding labels y_train
# Assume we have test data X_test and corresponding labels y_test

# Initialize KNN classifier with k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Train the KNN classifier on the training data
knn.fit(X_train, y_train)

# Predict the labels of the test data using the trained KNN classifier
y_pred = knn.predict(X_test)

# Calculate the accuracy of the predictions on the test data
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```
